## Как открыть текстовый файл? 

Мы можем считать одиг простой текстовый файл в память, используя следующие шаги

````python
file_path = '/content/drive/MyDrive/Colab Notebooks/Python: text analysis/henry-iv-part-1_TXT_FolgerShakespeare.txt'
data = open(file_path) #open file
contents = data.read() #create object
data.close() #close file

print(contents[:100])
````

Мы открываем объект файла для доступа к содержимому обычной текстовой версии одной из пьес Шекспира: 
1.   Местоположение этого файла указывается в качестве единственного аргумента функции open(). 
2.   После загрузки файла фактическое содержимое файла считывается как строковый объект с помощью метода read().
3. Распечатка первых 100 символов показывает, что мы действительно получили ряд символов, читаемых человеком.
4. Очень важно, что соединения с файлами должны быть закрыты, как только они больше не нужны: вызов close() гарантирует, что поток данных к исходному файлу будет прерван. 

Но лучше использовать конструкцию **with open(file) as f:**


````python
with open(file_path) as data:
  contents = data.read()

print(contents[:100])
````

Смысл конструкции **with** заключается в том, что он выполнит весь код, находящийся в его области видимости; однако после выполнения он закроет файл, даже если при чтении файла могла возникнуть ошибка.

Немного про кодировку: в настоящее время мир компьютерных технологий все больше полагается на так называемый стандарт Unicode, который охватывает более 128 000 символов. Стандарт Unicode реализован в различных стандартах кодирования, таких как UTF-8 и UTF-16. Стандарт UTF-8 стал стандартом кодирования текста и используется по умолчанию.

В некоторых случаях может понадобиться указать дополнительный аргумент кодировки в open() и убедиться, что вы загружаете файл с использованием правильной кодировки (например, open(..., encoding='utf8')).

## HTML

### Немного про язык html

**HTML (HyperText Markup Language)** – язык разметки (для тех, кто работал с *RMarkdown*: *Markdown* ‒  тоже язык разметки, только попроще). Поэтому, любая html-страница в интернете – это обычный текст, в котором отмечено, какие части представляют собой заголовок, абзац, таблицу, картинку и так далее. Вообще, правилом хорошего тона считается включать в html-код только содержательные вещи с разметкой, а все оформление и интерактив выносить в отдельные файлы.

Обычно, за страницей в интернете кроется примерно следующее. На сервере лежит несколько папок. В одной папке лежат сами html-страницы с размеченным текстом. В другой – хранятся файлы *css* (*Cascading Style Sheets*), в которых заданы параметры оформления разных страниц (цвет заголовков, фона, меню, ширина рамочек вокруг текста и картинок и прочее). В третьей папке находятся файлы с кодом на *JavaScript*, которые отвечают за всевозможный интерактив: будь то всплывающий перевод слова при наведении курсора, увеличение картинки при просмотре, подсветка формы для заполнения, если формат даты неверный, и так далее. JavaScript ‒ тоже отдельный язык программирования (несмотря на название, с Java не связан), на котором пишется код, соответствующий интерактиву.

Все эти файлы связаны между собой: к каждому html-файлу присоединены с помощью ссылок css-файл с оформлением и код javascript. Иногда, к сожалению, html-страницы создаются не по стандартам, и в файле с текстом встречаются огромные куски, отвечающие за оформление, что значительно затрудняет работу (на примеры таких сайтов мы потом посмотрим).

Сегодня мы начнем разбирать веб-скрейпинг и парсинг html-файлов. Обычно два этих процесса связаны, просто парсинг – более узкое понятие. Веб-скрейпинг (*web scraping*) – это вообще выгрузка данных с веб-страниц, а парсинг (*parsing*) – это выбор текста из файла с разметкой, например, из html или xml. Для разметки используются тэги – служебные слова в треугольных скобках `<>`. Тэги бывают разными: открывающие и закрывающие. Закрывающие тэги начинаются с прямого слэша (`/`). 

Вся страница заключается в тэг `<html></html>`. В начале обычно указывается какая-то мета-информация: язык страницы, кодировка, метки, название и прочее. Здесь, правда, такого нет. "Тело" документа, собственно страница, которая отображается при просмотре, заключается в тэг `<body></body>`. 

Для заголовков используются тэги с `h` (`h` – от *header*). Тэги `<h1></h1>`  ‒ для заголовков первого уровня, тэг `<h2></h2>` – для заголовков уровня, и так далее. В тэги `<p></p>` заключаются абзацы (`p` – от *paragraph*). Давайте для начала внесем изменения в файл ‒ создадим страничку о себе. ## JSON

**JavaScript Object Notation** - это легкий формат данных для хранения и обмена данными. Популярность JSON отчасти объясняется его лаконичным синтаксисом, который опирается на JavaScript и другие популярные языки программирования. JSON хранит информацию, используя четыре основных типа данных:

1. **Строка** (в двойных кавычках) string
2. **Число**(похоже на float в Python) number
3. **Логическое выражение** (true или false) boolean
4. **Null** (похоже на None в Python) null

А также две структуры данных для коллекций данных: 
1. Объект (object), который представляет собой коллекцию пар имя/значение, похожую на **dict **в Python, 
2. Массив (array), который представляет собой упорядоченный список значений, похожий на **list** в Python.


---

Давайте сначала рассмотрим **объекты**. Они заключаются в фигурные скобки ({}) и состоят из пар имя/значение, разделенных запятыми. Пары имя/значение в JSON очень похожи на синтаксис словаря Python, поскольку они имеют форму name: value. Как показано в следующем фрагменте JSON, имена представлены в виде строк в двойных кавычках:



````python
{
  "line_id": 14, #string, number and object
  "play_name": "Henry IV",
  "speech_number": 1,
  "line_number": "1.1.11",
  "speaker": "KING HENRY IV",
  "text_entry": "All of one nature, of one substance bred,"
}
````

Так же, как словари в Python отличаются от списков, объекты JSON отличаются от массивов. В массивах JSON используется тот же синтаксис, что и в Python: квадратные скобки с элементами, разделенными запятыми. Вот пример:

````python
[
  {
    "line_id": 12664,
    "play_name": "Alls well that ends well",
    "speech_number": 1,
    "line_number": "1.1.1",
    "speaker": "COUNTESS",
    "text_entry": "In delivering my son from me, I bury a second husband."
  },
  {
    "line_id": 12665,
    "play_name": "Alls well that ends well",
    "speech_number": 2,
    "line_number": "1.1.2",
    "speaker": "BERTRAM",
    "text_entry": "And I in going, madam, weep o'er my father's death"
  }
]
````

Важно отметить, что значения, в свою очередь, могут быть объектами или массивами, что позволяет создавать вложенные структуры. Как видно из приведенного примера, разработчики могут свободно смешивать и вкладывать массивы и объекты, содержащие пары имя-значение в JSON.

Модуль json в Python предоставляет ряд функций для удобного кодирования и декодирования объектов JSON. Кодирование объектов Python или иерархий объектов в виде строк JSON можно выполнить с помощью json.dumps()

Задание 1. Написать функцию, которая принимает список словарей и возвращает словарь с общими ключами и значениями

````python
def get_common_values(dict_list): 
  result = {} 
  for d in dict_list: 
    for k,v in d.items(): 
      if k in result: 
        if result[k] != v: 
          result[k] = None
        else: result[k] = v 
  return result
````

Задание 2. Написать функцию, которая вернет списком все значения для одного ключа

````python
def get_values_for_key(dict_list,key): 
  result = [] 
  for d in dict_list: 
    if key in d: 
      result.append(d[key]) 
  return result
````

Задание 3. Написать функцию, которая посчитает количество уникальных значений в списке

````python
def count_unique(lst): 
  unique = [] 
  for x in lst: 
    if x not in unique: 
      unique.append(x) 
  return len(unique)
````

Задание 4. Написать функцию, которая вернет словарь с указанием числа вхождений каждого элемента списка

````python
def count_elements(lst): 
  result = {} 
  for x in lst: 
    if x in result: 
      result[x] += 1 
    else: result[x] = 1 
  return result
````
